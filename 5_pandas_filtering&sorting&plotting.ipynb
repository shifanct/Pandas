# Plotting

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/data - data.csv")
df
```

```python
df.plot()
plt.show()
```

```python
#we can plot the Duration column only
plt.figure(figsize=(8,5))
plt.plot(df['Duration'],color='red',marker='*')
plt.title('Line Plot Of Duration')
plt.xlabel('index')
plt.ylabel('Duration')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df["Calories"],color="green",marker="x")
plt.title("Line plot of Calories")
plt.xlabel("index")
plt.ylabel("Calories")
plt.show()
```

## Sactter plot

```python
import matplotlib.pyplot as plt
# creating a scatter plot for 'calories' against the index of the dataset
plt.figure(figsize=(8,5))
plt.scatter(df.index, df["Pulse"],color="blue",marker="x")
plt.title("Scatter plot of pulse")
plt.xlabel("index")
plt.ylabel( "pulse")
# plt.grid(True)
plt.show()
```
