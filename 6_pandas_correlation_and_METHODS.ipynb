#49.combine()

```python
import pandas as pd

df1 = pd.DataFrame([[1,2,3],[13,4,3]])
df2 = pd.DataFrame([[5,6,7],[7,8,6]])
print(df1)
print("\n",df2)
```

```python
def myfunc(a,b):
   if (a.sum() > b.sum()):
     return a
   else:
     return b
```

```python
print(df2.combine(df1,myfunc))
```
